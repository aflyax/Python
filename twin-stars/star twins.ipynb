### Background

The functions in this notebook find, using random simulations, the expected probability of finding two twins stars. The numbers used (in the first and second functions) are 500 stars on a square of photo paper 1 cm x 1 cm.

Statistical theory yields the probability of finding two stars within 1 mm of each other on the above field equal to 0.146. This number can be calculated/corroborated using Monte Carlo simulation using a large (enough) number of trials, in each of which N stars are randomly generated on a square field of given dimensions, and the frequency of twin stars (points within a minimum distance of each other) is computed.

Knowing the expected probability is important, since the actual observed frequency of twin stars (>0.8) is much higher than expected, suggesting that there is some physical mechanism contributing to star clustering.

- <i>find_nbrs()</i> performs a single simulation and displays (by default) the figure of stars, including (highlighted) twins.

- <i>find_prob()</i> repeats the simulation over N trials and displays a graph of individual trials' probabilities and the updated mean probability (as of each trial).

- <i>enter_star_info()</i> allows to enter the number of stars, the dimensions of the field, the minimal distance, and the number of iterations interactively.

Future directions: compute the probability of having twin stars in 3D (using x, y, z coordinates) against the probability of the same stars being detected as twin when projected on a 2D surface (the latter should be higher, since two stars could be close to each other on the x-y plane but far on the z axis).

The idea is taken from <a href="http://www.amazon.com/Developing-Analytic-Talent-Becoming-Scientist/dp/1118810082"><i>Developing Analytic Talent</i></a> by Vincent Granville.

```python
%pylab inline
import pandas as pd
import seaborn as sns
```

```python
def find_nbrs(low, high, num, min_d, make_fig=True):
    # finds twin stars (min_d distance from each other) among random num stars on high x high field
    
    sns.set(context = 'poster', style = 'dark')

    # populating stars
    x = random.uniform(low, high, num)
    y = random.uniform(low, high, num)
    points = pd.DataFrame({'x':x, 'y':y})

    tot_twins = 0
    twins = []
    
    #go through each point in the DataFrame and compare its distance to every other point against min_d:
    for i in arange(len(points)):
        x_0 = points.x[i]                 #the coordinates of the point in question
        y_0 = points.y[i]

        twins = ( points[((x_0 - points.x)**2 + (y_0 - points.y)**2) < min_d**2] )        # twins = DataFrame containing all pairs
 
        if len (twins) > 1:     #needs to be done b/c pandas compares each point against itself (obviously finding the distance < min_d)
            tot_twins += 1
            if(make_fig): plot (twins.x, twins.y, 'or-') #highlight the stars id'ed as twins
    
    if(make_fig):                               # I hate the repeat of the if, but too lazy to figure out a more elegant way
        plot (points.x, points.y, 'b.')         # plot all the stars
    
    return tot_twins

# print find_nbrs(0, 100, 500, 1) / 500.
```

```python
def find_prob(num_stars, dim, min_d, num_iter):
    #iterates through num_iter trials and displays both P(twins) for each trial and the mean(P) over all trials as of the current trial
    
    sns.set(context = 'poster', style = 'dark')

    tot_twins_prob = []
    up_twins_prob = []

    figure()
    hold = True

    for i in arange (num_iter):
        tot_twins_prob.append(float(find_nbrs(0, 100, num_stars, 1, make_fig = False)) / num_stars) #don't wanna make fig for every trial
        up_twins_prob.append(mean(tot_twins_prob))
    
    figure()
    find_nbrs(0, dim, num_stars, min_d) # figure for rep purposes
    
    figure()
    plot (tot_twins_prob, '-b', label = 'individual trials')
    plot (up_twins_prob, '-r', label = 'grand mean')
    plot (range(num_iter), [0.146]*num_iter, '-.g', label = 'expected from theory')
    legend(loc = 'best')
    xlabel('trial')
    ylabel('probability (twin stars)')
    
#     print up_twins_prob[-1:]

# find_prob(500, 100, 1, 10)
```

```python
def enter_star_info():
    #enter conditions interactively
    
    num_stars = input('\nPlease enter the total number of stars: ')
    dim = input('Please enter one of the dimensions of the (square) field of view: ')
    min_d = input('Please enter the minimum distance for two stars to be considered twins: ')
    num_iter = input('Please enter the number of iterations: ')

    find_prob(num_stars, dim, min_d, num_iter)
    
    import winsound
    winsound.MessageBeep()          # for long computations with large num_iter
    
enter_star_info()
```
