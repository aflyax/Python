```python
#setting up the environment

%pylab inline
import pandas as pd

import seaborn as sns
sns.set(context = 'poster', style = 'dark')
```

```python
#importing providers_info

providers_info = pd.read_csv('providers.csv')
providers_info = providers_info[0:2884]                          # removing NA values at the end
providers_info.columns = ['practice_id', 'provider_id', 'specialty', 'provider_level', 'tenure']

key_id = providers_info.practice_id.astype(str) + providers_info.provider_id.astype(str)

providers_info['key_id'] = key_id

providers_info_nested = providers_info.set_index(['practice_id', 'provider_id', 'specialty', 'provider_level', 'tenure', 'key_id'])
# providers_info_nested.

providers_info.head(50)

# providers_info[providers_info.key_id == '1.01.0'].specialty

# providers_info.plot(x = 'specialty', y = 'tenure')
```

```python
#importing work_data

work_data = pd.read_csv('work_data.csv')
work_data = work_data[0:29974]
work_data.columns = ['practice_id', 'provider_id', 'obs_month', 'variable', 'value']

key_id = work_data.practice_id.astype(str) + work_data.provider_id.astype(str)
work_data['key_id'] = key_id
work_data_nested = work_data.set_index(['practice_id', 'provider_id', 'obs_month', 'variable', 'key_id']).unstack('variable')
work_data = pd.DataFrame(work_data_nested.reset_index())
# work_data = work_data.set_index(['key_id'])
work_data = work_data.rename(columns = {'value': 'work_vals'})

# global_df = work_data

# global_df.tail(10)
```

```python
#importing emr_touches and adding to global_df

emr_touches = pd.read_csv('emr_touches.csv')
emr_touches = emr_touches[0:14987]

emr_touches.columns = ['practice_id', 'provider_id', 'obs_month', 'keystrokes', 'mousestrokes']
key_id = emr_touches.practice_id.astype(str) + emr_touches.provider_id.astype(str)
emr_touches['key_id'] = key_id

# emr_touches_nested = emr_touches.set_index(['practice_id', 'provider_id', 'obs_month', 'key_id', 'keystrokes', 'mousestrokes'])

# emr_touches_nested.tail(10)

emr_touches.tail(10)

global_df = emr_touches
global_df['encounters'] = work_data.work_vals.Encounters
global_df['RVUs'] = work_data.work_vals.RVUs

global_df.tail(10)
```

```python
#importing emr_time

emr_time = pd.read_csv('emr_time.csv')
emr_time = emr_time[0:59948]

emr_time.columns = ['practice_id', 'provider_id', 'obs_month', 'variable', 'value']
key_id = emr_time.practice_id.astype(str) + emr_time.provider_id.astype(str)

emr_time['key_id'] = key_id
emr_time_nested = emr_time.set_index(['practice_id', 'provider_id', 'obs_month', 'variable', 'key_id']).unstack('variable')

emr_time = pd.DataFrame(emr_time_nested.reset_index())
# # work_data = work_data.set_index(['key_id'])
emr_time = emr_time.rename(columns = {'value': 'EMR_time'})

global_df['exam_emr'] = emr_time.EMR_time['Time Exam']
global_df['intake_emr'] = emr_time.EMR_time['Time Intake']
global_df['postvisit_emr'] = emr_time.EMR_time['Time Postvisit']
global_df['signoff_emr'] = emr_time.EMR_time['Time Signoff']

global_df.head(10)

#replace zero values with NAN

# print 'average ', global_df.intake_emr.mean()

# global_df.intake_emr = global_df.intake_emr.replace(0, nan)  ## do later if necc

# print 'average ', global_df.intake_emr.mean() #works -- mean doesn't take nan into account

global_df_ids = global_df

# #drop ids... for now

# global_df = global_df.drop('practice_id', 1)
# global_df = global_df.drop('provider_id', 1)

global_df.tail(10)
```
