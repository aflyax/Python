```python
def pop_means(total_df):
    total_df['exam_mee'] = total_df.exam_emr / total_df.encounters
    total_df['intake_mee'] = total_df.intake_emr / total_df.encounters
    total_df['postvisit_mee'] = total_df.postvisit_emr / total_df.encounters
    total_df['signoff_mee'] = total_df.signoff_emr / total_df.encounters
    total_df['RVUs_enc'] = total_df.RVUs / total_df.encounters
    total_df = total_df.rename(columns = {'mee':'grand_mee'})
    
    return total_df
    
total_df.columns
```
