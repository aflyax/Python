```python
def extract_fft(wav):
    import scipy as sp
    fft = abs(sp.fft(wav)[:1000])
    return fft
```

```python
def calc_mfcc(input, nwin=256, nfft=512, fs=16000, nceps=13):
    # MFCC parameters: taken from auditory toolbox
    over = nwin - 160
    # Pre-emphasis factor (to take into account the -6dB/octave rolloff of the
    # radiation at the lips level)
    prefac = 0.97

    #lowfreq = 400 / 3.
    lowfreq = 133.33
    #highfreq = 6855.4976
    linsc = 200/3.
    logsc = 1.0711703

    nlinfil = 13
    nlogfil = 27
    nfil = nlinfil + nlogfil

    w = hamming(nwin, sym=0)

    fbank = trfbank(fs, nfft, lowfreq, linsc, logsc, nlinfil, nlogfil)[0]

    #------------------
    # Compute the MFCC
    #------------------
    extract = preemp(input, prefac)
    framed = segment_axis(extract, nwin, over) * w

    # Compute the spectrum magnitude
    spec = np.abs(fft(framed, nfft, axis=-1))
    # Filter the spectrum through the triangle filterbank
    mspec = np.log10(np.dot(spec, fbank.T))
    # Use the DCT to 'compress' the coefficients (spectrum -> cepstrum domain)
    ceps = dct(mspec, type=2, norm='ortho', axis=-1)[:, :nceps]

    return ceps, mspec, spec
```

```python
# from scikits.talkbox.features import mfcc

def extract_mfcc(wav):
    import numpy as np
    ceps, _, _ = calc_mfcc(wav)
    num_ceps = ceps.shape[0]
    ceps[np.isinf(ceps)] = ceps.max()
    ceps[np.isneginf(ceps)] = ceps.min()
    ceps[np.isnan(ceps)] = 0.
    return np.nan_to_num(ceps[int(num_ceps*0.1):int(num_ceps*0.9)], )
```

```python
fft_X = np.load('fft.npy')
# fft_X.shape
train = fft_X
```

```python
labels = np.load('labels.npy')
labels.shape

# encode labels 
lbl_enc = preprocessing.LabelEncoder()
labels = lbl_enc.fit_transform(labels)
```

```python
from sklearn.preprocessing import StandardScaler

# train = fft_X

X_scaler = StandardScaler()

train_scaled = X_scaler.fit_transform(train)
# test_scaled = X_scaler.fit_transform(test)

train_scaled
```

```python
from sklearn.decomposition import RandomizedPCA

pca = RandomizedPCA(n_components=30, whiten=False)
pca.fit(train)
# plt.plot(pca.components_[0,:])
train_fac = pca.transform(train_scaled)
# test_fac = pca.transform(test_scaled)
np.shape(train_fac)

train_pca = np.concatenate((train, train_fac), axis=1)
# test_pca = np.concatenate((test, test_fac), axis=1)
np.shape(train_pca)
```
