```python
df = unemp_df.dropna()
df.reset_index(inplace=True, drop=True)
# df['region'] = ""


for i in df.index:
    for key in regions.keys():
        if df.ix[i].country in regions[key]:
#             print(df.ix[i].Country, " ", key)
            df.loc[i,"region"] = key
```

```python
# Define some CSS to control our custom labels
css = """
table
{
  border-collapse: collapse;
}
th
{
  color: #ffffff;
  background-color: #000000;
}
td
{
  background-color: #cccccc;
}
table, th, td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px solid black;
  text-align: right;
}
"""

fig, ax = plt.subplots()
ax.grid(True, alpha=0.3)

points = ax.plot(df.min_wage_month, df.youth_unemp, '.', )

plt.xlim([-100, 4500]);
plt.ylim([-1, 80]);
plt.xlabel("minimum wage (per month), USD")
plt.ylabel("youth unemployment, %")

labels = []
for i in range(len(df)):
    label = df.ix[[i], :].T
    label.columns = ['Row {0}'.format(i)]
    # .to_html() is unicode; so make leading 'u' go away with str()
    labels.append(str(label.to_html()))


tooltip = plugins.PointHTMLTooltip(points[0], labels,
                                   voffset=10, hoffset=10, css=css)
plugins.connect(fig, tooltip)
plt.title("the world (hover over points for info)")
mpld3.display()
```
